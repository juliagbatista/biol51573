### Get the length of a string

```python
# the long (but readable) way
name_length = len(full_name)
print(name_length)

# a more concise way to do it, functions within functions
print(len(full_name)) #dont need to worried about storing the length of the name in a variable, if we just want to print it out, we can just put the function within the print function
```

### Print strings and numbers

```python
# print strings and integers
print("There are " + str(name_length) + " characters in Beyonce's name.")
print("There are", str(name_length), "characters in Beyonce's name.") # this is a more concise way to do it, we can just put the function within the print function, and separate the different parts of the string with commas, which will automatically add a space between them

# formatted print
print(f"There are {name_length} characters in Beyonce's name.") # this is the most concise way to do it, we can just put the function within the print function, and use f-strings to format the string, which will automatically replace the variable with its value

# print strings and integers with formatted print
```

## Some operations on strings

```python
# methods to change the case of a string
print(first, last)
print(first.upper()) # make the name uppercase #name of the method is upper, and we use parentheses to call the method, which will return a new string with all the characters in uppercase
print(first.lower()) # make the name lowercase

#'type()' returns what type of variable something is 
print(type(first)) 

# make the name properly cased (play with this string to make sure our algorithm is robust)
db_name = 'AndY'

# can give a literal string to 'capitalize()'

print("andy".capitalize()) # we can also just use the capitalize method on the original string, which will give us the same result

#utimately, we want to make sure that our algorithm is robust, and can handle any input, so we can use the lower method to make all the characters lowercase, and then use the capitalize method to make the first character uppercase and the rest lowercase, which will give us the properly cased name
db_name_lower = db_name.lower() # we can also store the lowercase version of the name in a variable, and then use the capitalize method on that variable, which will give us the same result
db_name_cap = db_name_lower.capitalize() # we can also store the capitalized version of the name in a variable, and then print that variable, which will give us the same result
print(db_name_cap)

#here we do it all in one statement 
print(db_name.lower().capitalize()) # we can use the lower method to make all the characters lowercase, and then use the capitalize method to make the first character uppercase and the rest lowercase, which will give us the properly cased name
```

### Problem #1: Write a program that asks for the user's name and age, then prints a personalized greeting using string formatting (`.format`).

For *Andy* and *32*, the output would be:

Hello, Andy! You are 32 years old.


```python
# code here
name = input('enter your name: ') # the input function will prompt the user to enter their name, and then store that input in the variable 'name'
age = input('enter your age: ') # the input function will prompt the user to enter their age, and then store that input in the variable 'age'

name = name.lower().capitalize() 

print('Hello, {}! You are {} years old.'.format(name, age)) 
print(f'Hello, {name}! You are {age} years old.') 

```

### Problem #2: Ask the user for a big number. If the user enters 1000000 (one million), the output will be:

This sure is a big number: 1,000,000

This sure is a big number: 1.000000e+06

```python
big_num = input('Enter a big number: ')

#show that the result from input() is always a string
#print(type(big_num)) 

print(f'This sure is a big number: {int(big_num):,d}') # we can use the int() function to convert the string to an integer, and then use the format specifier ',d' to format the number with commas as thousands separators, which will make it easier to read
print(f'This sure is a big number: {int(big_num):,e}') # we can also use the format specifier ',e' to format the number in scientific notation, which will make it easier to read if the number is very large
```
