## Working with lists

```python
# declare an empty string
bob = ''

# declare a list
beatles = []
```

### What kind of variable is this?

```python
print(type(bob))
print(type(beatles))
```

### Populate and print a list

```python
beatles = ["John", "Paul", "George", "Ringo", "Stuart"]
#print the whole list
print(beatles)
```

### Calculate and print the length of a list

```python
# calculate number of beatles
number_of_beatles = len(beatles)

# print number of beatles
print(f"There were {number_of_beatles} members of the Beatles.")
```

### Print an item in a list 
   
   _see PowerPoint slides_

```python
#remind ourselves of the list
print(beatles)
# print the first beatle (John)
print(beatles[0])

# print the last Beatle
print(beatles[-1])
print(beatles[4])
print(beatles[len(beatles)-1])
print(beatles[number_of_beatles-1])

#print george
print(beatles[2])
print(beatles[-3])
print(beatles[len(beatles)-3])
print(beatles[number_of_beatles-3])
```

### Remove an item from a list

```python
# Remove the last item in the list (bye, Stuart)
stu = beatles.pop() #we save the popped item in a variable called stu

# What happened to the Beatles?
print(beatles)
print(stu)
```

### Make a new list

```python
# make a list called 'stones' with the Rolling Stones (Mick, Ron, and Keith)
stones = ["Mick", "Ron", "Keith"]

#print the list of Rolling Stones
print(stones)
```

### Add two lists

```python
# add two grops
print("the two bands separate: {},{}".format(beatles, stones))
# add two groups together
supergroup = beatles + stones

#print the supergroup
print(f"{len(supergroup)} members of the new band: {supergroup}")
```

### Growing a list with `extend()`

```python
# print our lists
print("Jackson 5:", jackson5)
print("Destiny's Child:", destiny)

```

```python

# use extend() to add Destiny's Child to the Jackson 5
# extend() modifies a list in place, it doesn't return a value (i.e., a new list)
jackson5.extend(destiny)

#here is what is doing 
#jackson5 = jackson5 + destiny

print(f"Here's the extended Jackson 5: {jackson5}")

```

### Growing a list with `append()`

```python
# use `append()` to add lists in a different way
num = [1, 2, 3]
let = ['a', 'b', 'c']

print(f"our lists: {let, num}")

#append let onto num 
num.append(let)



# What did append do?
print(f"after appending let onto num: {num}")

more_num = [4, 5, 6]
more_num.append(7)
print(more_num)
i=8
more_num.append(i)
print(more_num)
print(num)
```
